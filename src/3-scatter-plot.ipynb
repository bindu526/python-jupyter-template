### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)

for i, j in zip(x, y):
    plt.annotate(f"({i},{j})", (i, j), textcoords="offset points", xytext=(5, 5))

plt.title("Scatter Plot with Annotations")

plt.show()
```
